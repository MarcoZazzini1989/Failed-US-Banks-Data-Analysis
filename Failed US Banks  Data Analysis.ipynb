```python
df = df.fillna('')
```

```python
# Convert Closing_Date to datetime
df['Closing_Date'] = pd.to_datetime(df['Closing_Date'])
```

```python
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
sns.countplot(data=df, x='State', palette='viridis')

plt.title('Distribution of Failed Banks by State', fontsize=16, fontweight='bold')
plt.xlabel('State', fontsize=14)
plt.ylabel('Number of Failed Banks', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Adding data labels on top of each bar
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='gray', xytext=(0, 5),
                textcoords='offset points')

# Removing top and right spines
sns.despine()

plt.show()
```

```python

# Analyze the relationship between Estimated_Loss and Total_Assets


plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Customize the color palette
palette = sns.color_palette("Set1", n_colors=len(df['State'].unique()))

# Create the scatter plot
sns.scatterplot(data=df, x='Total_Assets', y='Estimated_Loss', hue='State', palette=palette, s=100)

plt.title('Estimated Loss vs. Total Assets', fontsize=16, fontweight='bold')
plt.xlabel('Total Assets', fontsize=14)
plt.ylabel('Estimated Loss', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Customize legend
legend = plt.legend(title='State', fontsize=12, title_fontsize=14,loc='lower left',bbox_to_anchor=(1, 0),ncol=4)
legend.get_title().set_fontweight('bold')






plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Customize the color palette
palette = sns.color_palette("coolwarm", n_colors=len(df['State'].unique()))

# Create the bar plot
sns.barplot(data=df, x='State', y='Estimated_Loss', ci=None, palette=palette)

plt.title('Average Estimated Loss by State', fontsize=16, fontweight='bold')
plt.xlabel('State', fontsize=14)
plt.ylabel('Average Estimated Loss', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.style.use('ggplot')

# Create the histogram plot with KDE
plt.hist(  x=df['Estimated_Loss'], bins=20,color='skyblue')

plt.title('Distribution of Estimated Loss', fontsize=16, fontweight='bold')
plt.xlabel('Estimated Loss', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Grouping and aggregating the data
df['Year'] = df['Closing_Date'].dt.year
closure_counts = df.groupby('Year').size()

# Create the line plot
sns.lineplot(data=closure_counts, marker='o', color='steelblue')

plt.title('Bank Closures Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Bank Closures', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()

plt.show()
```

```python
corr_matrix = df[['Estimated_Loss', 'Total_Assets', 'Total_Deposits']].corr()

# Create a diverging color palette for the heatmap
cmap = sns.diverging_palette(220, 20, as_cmap=True)

plt.figure(figsize=(10, 8))

# Create the heatmap with custom annotations and aesthetics
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap=cmap, center=0,
            linewidths=.5, annot_kws={"size": 12}, cbar_kws={"shrink": 0.8})

plt.title('Correlation Heatmap', fontsize=16, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
# Selecting a subset of numerical features for the pair plot
numerical_features = ['Estimated_Loss', 'Total_Assets', 'Total_Deposits']
pairplot_data = df[numerical_features]

# Customizing the pair plot
sns.set(style="ticks")
sns.pairplot(pairplot_data, diag_kind='kde', markers='o', plot_kws={'alpha': 0.7})

# Adding a title and adjusting plot layout
plt.suptitle('Pair Plot of Numerical Features', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()

plt.show()
```

```python
# Set the style for seaborn plots
sns.set(style="whitegrid")

# Create a function for creating and showing boxplots
def create_boxplot(data, x_col, y_col, title):
    plt.figure(figsize=(12, 8))
    ax = sns.boxplot(data=data, x=x_col, y=y_col)
    ax.set_title(title)
    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    plt.tight_layout()
    plt.show()

# Call the function for each boxplot
create_boxplot(df, 'State', 'Estimated_Loss', 'Distribution of Estimated Loss by State')
create_boxplot(df, 'State', 'Total_Assets', 'Distribution of Total Assets by State')
create_boxplot(df, 'State', 'Total_Deposits', 'Distribution of Total Deposits by State')
```
